### Data Cleaning

#### Cleaning column names, removing duplicates, and handling datatypes.

```python
gameplay.rename(columns = {'User ID': 'User Id'}, inplace = True)
```

```python
deposit = deposit.drop_duplicates(subset = ['User Id', 'Datetime', 'Amount']) 
```

```python
withdrawal.duplicated().sum()
```

```python
gameplay.duplicated().sum()
```

```python
gameplay = gameplay.drop_duplicates(subset = ['User Id', 'Games Played', 'Datetime'])
```

```python
gameplay.duplicated().sum()
```

```python

deposit['Datetime'] = pd.to_datetime(deposit['Datetime'], errors = 'coerce')
withdrawal['Datetime'] = pd.to_datetime(withdrawal['Datetime'], errors = 'coerce')
gameplay['Datetime'] = pd.to_datetime(gameplay['Datetime'], errors = 'coerce')
```

### Feature Engineering

```python
# created time slots (S1 and S2) 
# Extracted date from time stamp
```

```python
def get_slot(hour):
    if hour < 12:
        return "S1"
    else:
        "S2"
```

```python
# create slot 
deposit['Slot'] = deposit['Datetime'].dt.hour.apply(get_slot)
withdrawal['Slot'] = withdrawal['Datetime'].dt.hour.apply(get_slot)
gameplay['Slot'] = gameplay['Datetime'].dt.hour.apply(get_slot)
```

```python
# extract date
deposit['Date'] = deposit['Datetime'].dt.date
withdrawal['Date'] = withdrawal['Datetime'].dt.date
gameplay['Date'] = gameplay['Datetime'].dt.date
```

### Aggregation

```python
# Aggregatin data at user id, date, and slot level
```

```python
deposit_summary = deposit.groupby(['User Id', 'Date', 'Slot']).agg(deposit_amount = ('Amount', 'sum'), 
                                                                    deposit_count = ('Amount', 'count')
                                                                   ).reset_index()
```

```python
deposit_summary.head()
```

```python
withdrawal_summary = withdrawal.groupby(['User Id', 'Date', 'Slot']).agg(withdrawal_amount = ('Amount', 'sum'), 
                                                                         withdrawal_count = ('Amount', 'count')
                                                                        ).reset_index()
```

```python
withdrawal_summary.head()
```

```python
gameplay_summary = gameplay.groupby(['User Id', 'Date', 'Slot']).agg(games_played = ('User Id', 'count')
                                                                    ).reset_index()
```

```python
gameplay_summary.head()
```

### Data Merging

```python
# combining all datasets into one
```

```python
df = deposit_summary.merge(withdrawal_summary, on = ['User Id', 'Date', 'Slot'], how= 'outer')
```

```python
df.head()
```

```python
df = df.merge(gameplay_summary, on = ['User Id', 'Date', 'Slot'], how = 'outer')
```

```python
df.head(10)
```

```python
df.fillna(0, inplace= True)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
```

```python
df.info()
```

### Loyalty Points Calculation

```python
df['loyalty_points'] = (
    0.01 * df['deposit_amount'] + 
    0.005 * df['withdrawal_amount'] + 
    0.001 * np.maximum(df['deposit_count'] -df['withdrawal_count'], 0) + 
    0.2 * df['games_played']
)
```

### Q1. Slot wise Analysis

#### Loyalty points for specific dates and Slots
#### Calculating player wise loyalty points for given dates and slots.

```python
result_q1 = df[
    ((df['Date'] == '2022-10-02') & (df['Slot'] == 'S1')) |
    ((df['Date'] == '2022-10-16') & (df['Slot'] == 'S2')) |
    ((df['Date'] == '2022-10-18') & (df['Slot'] == 'S1')) |
    ((df['Date'] == '2022-10-26') & (df['Slot'] == 'S2'))
]
```

```python
result_q1.head()
```

### Q2. Monthly Leaderboard

```python
leaderboard = df.groupby('User Id').agg(total_points = ('loyalty_points', 'sum'), 
                                        total_games=('games_played', 'sum')
                                       ).reset_index()
```

```python
leaderboard = leaderboard.sort_values(by=['total_points', 'total_games'], ascending = False)
```

## Part B

### Bonus Allocation Strategy

After identifying the top 50 players based on loyalty points, a fair bonus distribution strategy is required for allocating ₹50,000.

#### Proposed Approach:

A hybrid model is recommended:

- 70% weight based on loyalty points (financial contribution)
- 30% weight based on number of games played (engagement)

#### Reasoning:

- Loyalty points capture both deposits and withdrawals
- Gameplay ensures active engagement is rewarded
- This prevents bias towards only high-spending users

#### Conclusion:

This hybrid approach ensures a balanced and fair distribution of bonus among top players.

## Part C

## Evaluation of Loyalty Formula

### Is the formula fair?

The current loyalty formula is partially fair but has some limitations:

- It favors users with higher deposits
- Players with high engagement but low deposits may receive fewer rewards

---

### Suggested Improvements:

1. Introduce a cap on deposit contribution to avoid dominance by high spenders  
2. Increase the weight of gameplay activity  
3. Normalize scores across users  
4. Add fraud detection mechanisms

The data was cleaned and analyzed properly, and loyalty points were calculated based on user activity. The analysis helped identify top players and showed that both gameplay and financial contribution impact user rewards. A balanced bonus strategy can further improve fairness and user engagement.
